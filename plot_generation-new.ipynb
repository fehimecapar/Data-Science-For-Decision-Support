```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm # Import colormaps

# 1. Create dummy data: 4 categories with 5 values each
np.random.seed(42) # for reproducibility
# Generating data with different ranges to make the boxplots distinct
data_a = np.random.rand(5)
data_b = np.random.rand(5)
data_c = np.random.rand(5)
data_d = np.random.rand(5)

data = [data_a, data_b, data_c, data_d]
category_labels = ['Logistic', 'Decision tree', 'SVM', 'kNN']

# 2. Create the boxplot
plt.figure(figsize=(8, 6))
# Using patch_artist=True to fill the boxes with color
bp = plt.boxplot(data, labels=category_labels, patch_artist=True, medianprops=dict(color='black'))

# Optional: Add colors to the boxes
colors = ['lightblue', 'lightgreen', 'pink', 'lightsalmon']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

# 3. Label the plot
plt.title('Accuracy (5 random splits)', fontsize=30)
plt.xlabel('Model', fontsize=25)
plt.ylabel('Accuracy', fontsize=25)

plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
# 4. Save the plot
plt.savefig('boxplot_4_categories.png')
plt.close()
```

```python
# --- 1. Generate Synthetic Data ---

np.random.seed(42)
# Generate 100 true observation values (y_true)
y_true = np.random.uniform(20, 100, 100)

# Simulate model predictions (y_pred) by adding some noise to y_true
# The noise ensures the points don't perfectly align, simulating real model error
noise = np.random.normal(0, 5, 100) # Mean 0, Standard Deviation 5
y_pred = y_true + noise

# --- 2. Create the Parity Plot ---

plt.figure(figsize=(7, 7))

# Scatter plot of True vs. Predicted values
plt.scatter(y_true, y_pred, alpha=0.7, color='navy', edgecolors='k', s=50)

# --- 3. Add the Parity Line (y = x) ---

# Determine the range for the parity line
min_val = min(y_true.min(), y_pred.min())
max_val = max(y_true.max(), y_pred.max())

# Create a straight line from (min_val, min_val) to (max_val, max_val)
plt.plot([min_val, max_val], [min_val, max_val], 
         color='red', 
         linestyle='--', 
         linewidth=2, 
         label='Parity Line ($y = \hat{y}$)')

# --- 4. Customize the Plot ---

plt.title('Parity Plot: Observed vs. Predicted Values', fontsize=14)
plt.xlabel('Observed (True) Values ($y$)', fontsize=12)
plt.ylabel('Predicted Values ($\hat{y}$)', fontsize=12)
plt.legend(loc='upper left')
plt.grid(True, linestyle=':', alpha=0.6)
plt.axis('equal') # Ensures the axes have the same scale
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm 
from matplotlib.patches import Patch 
import matplotlib as mpl

# --- 1. Data Generation (Reusing previous data setup) ---
n_categories = 4
category_labels = ['Logistic', 'Decision tree', 'SVM', 'kNN']

np.random.seed(42)
values_bar1=(0.8,0.9,0.7,0.6)#np.random.rand(n_categories)
# Values for the second bar in each category
#values_bar2 = np.random.randint(5, 25, n_categories) # e.g., sales for product type 2
values_bar2=(0.75,0.85,0.65,0.55)#np.random.rand(n_categories)

# --- 2. Plotting Setup ---
bar_width = 0.35
r = np.arange(n_categories) 
bar1_positions = r - bar_width/2
bar2_positions = r + bar_width/2

plt.figure(figsize=(12, 8)) 

# --- 3. Create the Grouped Bar Chart with different color palettes per category ---
category_colormaps = ['Blues', 'Greens', 'Oranges', 'Purples'] 

for i in range(n_categories):
    cmap = mpl.colormaps[category_colormaps[i]]
    color_bar1 = cmap(0.4) 
    color_bar2 = cmap(0.7) 
    
    # Plot bar 1 and add text annotation
    plt.bar(bar1_positions[i], values_bar1[i], color=color_bar1, width=bar_width, 
            edgecolor='grey')
    plt.text(bar1_positions[i], values_bar1[i] + 0.02, f'{round(values_bar1[i],2)}', ha='center', va='bottom', fontsize=12)

    # Plot bar 2 and add text annotation
    plt.bar(bar2_positions[i], values_bar2[i], color=color_bar2, width=bar_width, 
            edgecolor='grey')
    plt.text(bar2_positions[i], values_bar2[i] + 0.02, f'{round(values_bar2[i],2)}', ha='center', va='bottom', fontsize=12)


# --- 4. Customize the Plot ---
plt.xlabel('Models', fontsize=30)
plt.ylabel('Accuracy', fontsize=30)
plt.title('Accuracy of the models: random splits vc CV', fontsize=30)

plt.xticks(r, category_labels, fontsize=13) 
plt.tick_params(axis='y', labelsize=13)

# Create custom legend entries
legend_handles = [
    Patch(facecolor=mpl.colormaps['gray'](0.7), edgecolor='grey', label='Random splits'),
    Patch(facecolor=mpl.colormaps['gray'](0.4), edgecolor='grey', label='CV')
]
plt.legend(handles=legend_handles, loc='upper left', fontsize=13) 

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() 

# 5. Save the plot

plt.show()
plt.savefig('grouped_bar_chart_larger_font.png')
plt.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Define category labels and values (Original Order)
n_bars = 5
categories = [f'var {i+1}' for i in range(n_bars)] 
np.random.seed(42) 
values = np.random.randint(10, 50, n_bars)
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold', 'mediumpurple']

# 2. Create the bar plot
plt.figure(figsize=(8, 6))

# Plot without sorting
plt.bar(categories, values, color=colors, edgecolor='black')

# 3. Add labels and title
plt.title('Linear models parameters', fontsize=16)
plt.xlabel('Parameters', fontsize=14)
plt.ylabel('Value', fontsize=14)

# 4. Add value labels on top of the bars
for i, v in enumerate(values):
    plt.text(i, v + 0.5, str(v), color='black', ha='center', fontsize=12)

# 5. Add grid and adjust layout
plt.ylim(0, max(values) + 5) 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
# 6. Save the plot
plt.savefig('bar_plot_5_categories_unsorted.png')
plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the four example confusion matrices (rows are true labels, columns are predicted labels)
# Order: [[True Negative, False Positive], [False Negative, True Positive]]

# 1. Model A: Excellent performance
cm_a = np.array([[95, 5], [2, 98]])

# 2. Model B: High False Positives (overly sensitive)
cm_b = np.array([[80, 20], [5, 95]])

# 3. Model C: Poor/Near Random performance
cm_c = np.array([[55, 45], [40, 60]])

# 4. Model D: High False Negatives (low recall)
cm_d = np.array([[90, 10], [50, 50]])

all_cms = {
    'Logistic': cm_a,
    'Decision tree': cm_b,
    'SVM': cm_c,
    'kNN': cm_d
}

# Setup the 2x2 subplot figure
fig, axes = plt.subplots(2, 2, figsize=(7, 6))
axes = axes.flatten() # Flatten the 2x2 array of axes for easy iteration

# Labels for the confusion matrices
cm_labels = ['Negative', 'Positive']
cm_ticks = [0.5, 1.5] # Center ticks between 0 and 1, and 1 and 2

# Plot each confusion matrix
for i, (title, cm) in enumerate(all_cms.items()):
    ax = axes[i]
    
    # Create heatmap
    sns.heatmap(
        cm, 
        annot=True,          # Show the counts in the cells
        fmt='d',             # Format as integer (digit)
        cmap='Blues',        # Colormap for visual contrast
        cbar=False,          # Do not show a separate color bar
        linewidths=.5,       # Add lines between cells
        linecolor='black',
        annot_kws={"size": 18}, # INCREASED FONT SIZE FOR NUMBERS IN CELLS
        ax=ax
    )
    
    # Set titles and labels
    ax.set_title(title, fontsize=16)
    ax.set_xlabel('Predicted Label', fontsize=14)
    ax.set_ylabel('True Label', fontsize=14)
    
    # Set tick labels
    ax.set_xticks(cm_ticks)
    ax.set_xticklabels(cm_labels, fontsize=12)
    ax.set_yticks(cm_ticks)
    ax.set_yticklabels(cm_labels, rotation=90, va='center', fontsize=12)

# Adjust layout
plt.tight_layout()
plt.show()
plt.savefig('four_confusion_matrices_larger_cell_font.png')
plt.close()

print("four_confusion_matrices_larger_cell_font.png generated successfully.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Generate Synthetic Data (Now 5 lines, including 'Real Value') ---
np.random.seed(42)
N_POINTS = 10
x_values = np.linspace(0, 10, N_POINTS)

# Data for 4 models (same as previous step, but now representing predictions)
y_linear = 2 * x_values + np.random.normal(loc=0, scale=1.5, size=N_POINTS)
y_dtree = -0.5 * x_values + 15 + np.random.normal(loc=0, scale=1.5, size=N_POINTS)
y_svm = -3 * x_values + 30 + np.random.normal(loc=0, scale=1.5, size=N_POINTS)
y_knn = 0.5 * x_values + 5 + np.random.normal(loc=0, scale=1.5, size=N_POINTS)

# NEW: Real Value line - let's make it a clear, consistent trend without much noise
# We'll base it on one of the original trends, but make it distinct.
y_real_value = x_values * 1.5 + 10 # A clear, increasing trend for real values

# Define the line labels (now 5)
line_labels = ['Linear', 'Decision tree', 'SVM', 'kNN', 'Real Value']

# Group data for plotting
all_y_data = [y_linear, y_dtree, y_svm, y_knn, y_real_value]
colors = ['blue', 'red', 'green', 'purple', 'black'] # Black for Real Value
markers = ['o', 's', 'v', '^', 'X'] # 'X' for Real Value
x_labels = [f'observation {i+1}' for i in range(N_POINTS)]


# --- 2. Create the Plot (5 sets of points) ---
plt.figure(figsize=(10, 7))

# Plot all five sets of points
for y_data, label, color, marker in zip(all_y_data, line_labels, colors, markers):
    plt.plot(x_values, y_data, 
             label=label, 
             color=color, 
             linestyle='none', 
             marker=marker, 
             markersize=10, 
             linewidth=0)

# --- 3. Customize the Plot ---
plt.title('Model Predictions vs. Real Values', fontsize=16)
plt.xlabel('Observations from the testset', fontsize=20) 
plt.ylabel('Value', fontsize=20)

# Set custom X-axis ticks
plt.xticks(x_values, x_labels, rotation=45, ha='right', fontsize=12) 

plt.legend(title='Data / Model', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.tight_layout()

plt.show()
# 4. Save the plot
plt.savefig('models_and_real_values_points.png')
plt.close()

print("models_and_real_values_points.png generated successfully.")
```
